```python
srcs = []
labels = []
predicts = []
for k,v in data.items():
    srcs.append(v['src'])
    labels.append(v['label'])
    predicts.append(v['predict'])

```

```python
srcs = np.array(srcs)
labels = np.array(labels)
predicts = np.array(predicts)
```

```python
for k in range(len(srcs)):
    if labels[k] == predicts[k]:
        print("======>>>> CORRECT <<<<<=======")
        print("labels[k]", labels[k])
        print("input length", len(srcs[k].split()))
    # else:
    #     print("======>>>> WRONG <<<<<=======")
    #     print(f"labels[k]={labels[k]}, predicts[k]={predicts[k]}")
    #     print("input length", len(srcs[k].split()))
```

```python
corrects = []
wrongs = []
for k in range(len(srcs)):
    if labels[k] == predicts[k]:
        corrects.append(len(labels[k].split()))
        print("======>>>> CORRECT <<<<<=======")
        print("labels[k]", labels[k])
        print("input length", len(srcs[k].split()))
    else:
        wrongs.append(len(predicts[k].split()))
        print("======>>>> WRONG <<<<<=======")
        print(f"labels[k]={labels[k]}, predicts[k]={predicts[k]}")
        print("input length", len(srcs[k].split()))
```
